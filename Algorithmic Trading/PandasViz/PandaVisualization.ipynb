```python
idx=mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock=mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')

ax.xaxis.grid(True)
ax.yaxis.grid(True)

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')

ax.xaxis.grid(True)
ax.yaxis.grid(True)

#Locating
ax.xaxis.set_major_locator(dates.MonthLocator())

#Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n%b-%Y'))


ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
plt.tight_layout()
```
