```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers=['^GSPC','^IXIC','^GDAXI','^FCHI']

ind_data=pd.DataFrame()

for t in tickers:
    ind_data[t]=wb.DataReader(t,data_source='yahoo',start='1997-01-01')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data/ind_data.iloc[0]*100).plot(figsize=(12,8));
```

```python
ind_return=(ind_data/ind_data.shift(1))-1
ind_return
```

```python
ind_return.tail()
```

```python
ind_return.head()
```

```python
ind_return_an=ind_return.mean()*250
ind_return_an
```
