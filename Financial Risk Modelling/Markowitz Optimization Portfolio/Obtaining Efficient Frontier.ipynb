```python
# Expected Portfolio Return
expect_prtf_ret=np.sum(weights*log_returns.mean())*250
expect_prtf_ret
```

```python
# Expected Portfolio Variance
expect_prtf_varn=np.dot(weights.T,np.dot((log_returns.cov()*250),weights))
expect_prtf_varn
```

```python
# Expected Portfolio Volatility
expect_prtf_volat=np.sqrt(expect_prtf_varn)
expect_prtf_volat
```
