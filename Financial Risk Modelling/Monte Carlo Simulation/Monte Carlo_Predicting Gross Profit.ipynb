```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
rev_m=170
rev_stdev=20
iterations=1000
```

```python
rev=np.random.normal(rev_m,rev_stdev,iterations)
rev
```

```python
plt.figure(figsize=(12,8));
plt.plot(rev);
```

```python
COGS=-(rev*np.random.normal(0.6,0.1)) # Cost of Goods Sold
COGS
```

```python
plt.figure(figsize=(12,8));
plt.plot(COGS);
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
# Part II
```

```python
Gross_Profit=rev+COGS
Gross_Profit
```

```python
plt.figure(figsize=(12,8));
plt.plot(Gross_Profit);
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```
