```python
# Part I

# Euler's Method
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dy=lambda x,y: x*y # Numerical Solution

f=lambda x: np.exp(pow(x,2)/2) # Analytical Solution

x=0
xd=2
y=1
h=0.2
n=int((xd-x)/h)

xgr=np.linspace(x,xd,n+1)
ygr=np.empty(n+1,float)
ygr[0]=y

print('x \t\t y(Euler/h=0.1)\t f(x)(Analytical)')

for i in range(1,n+1):
    y+=dy(x,y)*h
    x+=h
    ygr[i]=y
    print('%f \t %f \t %f' %(x,y,f(x)))

plt.figure(figsize=(9,6))
plt.plot(xgr,ygr,'--r',xgr,f(xgr),'-k');
#plt.xlim(0,3);
#plt.ylim(0,10);
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['h=0.1','Analytical'])
plt.grid()
```

```python
# Improved Euler's Method

dy=lambda x,y: x*y # Numerical Solution

f=lambda x: np.exp(pow(x,2)/2) # Analytical Solution

x=0
xd=2
y=1
h=0.2
n=int((xd-x)/h)

xgr=np.linspace(x,xd,n+1)
ygr=np.empty(n+1,float)
ygr[0]=y

print('x \t\t y(EulerImprv/Num)\t f(x)(Analytical)')

for i in range(1,n+1):
    y1=y+(dy(x,y))*h
    y+=(dy(x,y)+dy(x+h,y1))*(h/2)
    x+=h
    ygr[i]=y
    print('%f \t %f \t %f' %(x,y,f(x)))

plt.figure(figsize=(9,6))
plt.plot(xgr,ygr,'--r',xgr,f(xgr),'-k');
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['h=0.1','Analytical'])
plt.grid()
```

```python
# Example 2: Euler's Method

dy=lambda x,y: -25*pow(x,2)+(5*y)+2

f=lambda x: 2*np.exp(x*5)+5*pow(x,2)+2*x # Analytical Solution

x=0
xd=1
y=2
h=0.1
n=int((xd-x)/h)

xgr=np.linspace(x,xd,n+1)
ygr=np.empty(n+1,float)
ygr[0]=y

print('x \t\t y(Euler/h=0.1)\t f(x)(Analytical)')

for i in range(1,n+1):
    y+=dy(x,y)*h
    x=i*h
    ygr[i]=y
    print('%f \t %f \t %f' %(x,y,f(x)))
    
plt.figure(figsize=(9,6))
plt.plot(xgr,ygr,'--r',xgr,f(xgr),'-k');
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['h=0.1','Analytical'])
plt.grid()
```

```python
# Part III

# Runge-Kutta Method: 2nd Order
```

```python
dy=lambda x,y: x*y # Numerical Solution

f=lambda x: np.exp(pow(x,2)/2) # Analytical Solution

# xgr=np.linspace(0,2,21)

x=0
y=1
xd=2
h=0.2
n=int((xd-x)/h)

xgr=np.linspace(x,xd,n+1)
ygr=np.empty(n+1,float)
ygr[0]=y

print('x \t\t y(Euler/Num)\t f(x)(Analytical)')

for i in range(1,n+1):
    k1=dy(x,y)*h
    k2=dy(x+(h/2),(y+0.5*k1))*h
    y+=k2
    x+=h
    ygr[i]=y
    
    print('%f \t %f \t %f' %(x,y,f(x)))

plt.figure(figsize=(9,6))
plt.plot(xgr,ygr,'--r',xgr,f(xgr),'-k');
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['RK4(h=0.1)','Analytical'])
plt.grid()
```

```python
# Part IV

# Runge-Kutta Method: 4th Order
```

```python
dy=lambda x,y: x*y # Numerical Solution

f=lambda x: np.exp(pow(x,2)/2) # Analytical Solution

x=0
y=1
xd=2
h=0.2
n=int((xd-x)/h)

xgr=np.linspace(x,xd,n+1)
ygr=np.empty(n+1,float)
ygr[0]=y

print('x \t\t y(Euler/Num)\t f(x)(Analytical)')

for i in range(1,n+1):
    k1=dy(x,y)*h
    k2=dy(x+(h/2),(y+0.5*k1))*h
    k3=dy(x+(h/2),(y+0.5*k2))*h
    k4=dy((x+h),(y+k3))*h
    y+=(1/6)*(k1+2*k2+2*k3+k4)
    x+=h
    ygr[i]=y
    
    print('%f \t %f \t %f' %(x,y,f(x)))
    
plt.figure(figsize=(9,6))
plt.plot(xgr,ygr,'--r',xgr,f(xgr),'-k');
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['RK4(h=0.1)','Analytical'])
plt.grid()
```
