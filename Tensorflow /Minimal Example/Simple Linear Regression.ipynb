```python
# import relevant libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import importlib
importlib.import_module('mpl_toolkits').__path__
#from mpl_toolkits.mplot3D import Axes3D
```

```python
# Part II
```

```python
# Generate random input data to train on
```

```python
observations=1000

xs=np.random.uniform(low=-10,high=10,size=(observations,1)) # variable 1
zs=np.random.uniform(low=-10,high=10,size=(observations,1)) # variable 1

inputs= np.column_stack((xs,zs))

print(inputs.shape) # n x k
```

```python
# Create target we are aiming at
```

```python
noise = np.random.uniform(-1,1,(observations,1))
targets = 2*xs - 3*zs + 5 + noise
print(targets.shape)
```

```python
# Part III
```

```python
# Plotting the Data
```

```python
# Goal: to see that there is a strong trend that our model learn to respond
```

```python
# For 3D plot, we are using reshape where we take as arguments the
# dimensions in which we want to fit the object.
targets = targets.reshape(observations,)

# Declaration of the figure:
fig =plt.figure()

#Cration of the 3D plot
ax=fig.add_subplot(111,projection='3d')

#Axes selections
ax.plot(xs,zs,targets)

#Labels
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_xlabel('Targets')

#plotting the data under different angle; hier :100
ax.view_init(azim=100)
plt.show()

# Reshaping the targets back to the shape that they were in before plotting
# This relationship is a side-effect of the 3D plot. 
targets = targets.reshape(observations,1)
```

```python
# Initialize the variables
```

```python
init_range=0.1

weights = np.random.uniform(-init_range,init_range,size=(2,1))

biases= np.random.uniform(-init_range,init_range,size=1)

print(weights)
print(biases)
```

```python
# Part IV
```

```python
# Set a learning Rate
```

```python
learning_rate=0.02
```

```python
# Train the model
```

```python
for i in range(100):
    outputs= np.dot(inputs,weights)+(biases)
    deltas=outputs-targets
    #loss=np.sum(deltas**2)/2/observations
    loss=(np.sum(pow(deltas,2)))/(2*observations)
    
    print(loss)
    
    deltas_scaled=deltas/observations
    
    weights=weights-learning_rate*np.dot(inputs.T,deltas_scaled)
    biases=biases-learning_rate*np.sum(deltas_scaled)
```

```python
print(weights,biases)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
