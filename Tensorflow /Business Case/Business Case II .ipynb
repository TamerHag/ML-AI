```python
import numpy as np
import tensorflow as tf
```

```python
# Data
```

```python
# Train data
npz=np.load('Audiobooks_data_train.npz')

train_inputs=npz['inputs'].astype(np.float)
train_targets=npz['targets'].astype(np.int)

# Validation data

npz=np.load('Audiobooks_data_validation.npz')
validation_inputs, validation_targets=npz['inputs'].astype(np.float), npz['targets'].astype(np.int)

# Test data
npz=np.load('Audiobooks_data_test.npz')

test_inputs=npz['inputs'].astype(np.float)
test_targets=npz['targets'].astype(np.int)
```

```python
# Model

# Outline, Optimizers, Loss, Early Stopping and Training
```

```python
input_size=10
output_size=2
hidden_layer_size=50

model=tf.keras.Sequential([
                        tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                        tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                        tf.keras.layers.Dense(output_size, activation='softmax')
                        ])
# For Classifier Model --> softmax method
```

```python
# The optimizer and the loss function
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Training
```

```python
batch_size=100

max_epochs=100
```

```python
model.fit(train_inputs,
          train_targets,
          batch_size=batch_size,
          epochs=max_epochs,
          verbose=2,
          validation_data=(validation_inputs,validation_targets)
         )
```

```python
batch_size=100

max_epochs=100

early_stopping=tf.keras.callbacks.EarlyStopping()
```

```python
model.fit(train_inputs,
          train_targets,
          batch_size=batch_size,
          epochs=max_epochs,
          verbose=2,
          callbacks=[early_stopping],
          validation_data=(validation_inputs,validation_targets)
         )
```

```python
early_stopping=tf.keras.callbacks.EarlyStopping(patience=2)

model.fit(train_inputs,
          train_targets,
          batch_size=batch_size,
          epochs=max_epochs,
          verbose=2,
          callbacks=[early_stopping],
          validation_data=(validation_inputs,validation_targets)
         )
```

```python
# Test the model
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets)
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets,verbose=2)
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets,batch_size=batch_size,verbose=2)
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets,batch_size=batch_size,verbose=2,callbacks=[early_stopping])
```
